# Package and Data

```python
import numpy as np
import pandas as pd
data = pd.read_csv("winequality.csv")
data
```

# Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
data["quality"] = pd.cut(data["quality"], bins = (2,6,8), labels = ["bad", "good"], right = True)
le = LabelEncoder()
data["quality"] = le.fit_transform(data["quality"])
data
```

```python
data["quality"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size = 0.3)
X_train = train.drop("quality", axis = 1)
X_test = test.drop("quality", axis = 1)
y_train = train["quality"]
y_test = test["quality"]
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(X_train)
X_train = pd.DataFrame(ss.transform(X_train), columns = X_train.columns)
ss = StandardScaler()
ss.fit(X_test)
X_test = pd.DataFrame(ss.transform(X_test), columns = X_test.columns)
```

```python
X_train
```

# Training and Testing

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
```

```python
from sklearn.neighbors import KNeighborsClassifier
knc_score = []
for i in range(10,501,10):
    knc = KNeighborsClassifier(n_neighbors = i)
    knc.fit(X_train, y_train)
    y_pred = knc.predict(X_test)
    knc_score.append(accuracy_score(y_test, y_pred))
```

```python
fig, ax = plt.subplots(1,figsize = (15,15))
for i in range(10,201,10):
    plt.text(i, knc_score[int(i/10-1)],i)
plt.plot([i for i in range(10,501,10)],knc_score)
plt.show()
```

```python
result = []

knc = KNeighborsClassifier(n_neighbors = 20)
knc.fit(X_train, y_train)
y_pred = knc.predict(X_test)

acc = accuracy_score(y_test, y_pred)
fl = f1_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
result.append(("KNeighborsClassifier with n_neighbors 20",acc,fl,prec,recall))
```

```python
pd.DataFrame(result, columns = ["model", "accuracy_score", "f1_score", "precision_score", "recall_score"])
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc_score = []
for i in range(1, 11):
    dtc = DecisionTreeClassifier(random_state = 0, max_depth = i)
    dtc.fit(X_train, y_train)
    y_pred = dtc.predict(X_test)
    dtc_score.append(accuracy_score(y_test, y_pred))
```

```python
fig, ax = plt.subplots(1,figsize = (15,15))
for i in range(1, 11):
    plt.text(i,dtc_score[i-1],i)
plt.plot([i for i in range(1, 11)],dtc_score)
plt.show()
```

```python
dtc = DecisionTreeClassifier(random_state = 0, max_depth = 4)
dtc.fit(X_train, y_train)
y_pred = dtc.predict(X_test)

acc = accuracy_score(y_test, y_pred)
fl = f1_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

result.append(("DecisionTreeClassifier with max_depth 4",acc,fl,prec,recall))
```

```python
pd.DataFrame(result, columns = ["model", "accuracy_score", "f1_score", "precision_score", "recall_score"])
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc_score = []
for i in range(10, 151, 10):
    rfc = RandomForestClassifier(random_state = 8, n_estimators = i)
    rfc.fit(X_train, y_train)
    y_pred = rfc.predict(X_test)
    rfc_score.append(accuracy_score(y_test, y_pred))  
```

```python
fig, ax = plt.subplots(1,figsize = (15,15))
for i in range(10, 151, 10):
    plt.text(i,rfc_score[int(i/10-1)],i)
plt.plot([i for i in range(10, 151, 10)],rfc_score)
plt.show()
```

```python
rfc = RandomForestClassifier(random_state = 0, n_estimators = 20)
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)

acc = accuracy_score(y_test, y_pred)
fl = f1_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

result.append(("RandomForestClassifier with n_estimators 20",acc,fl,prec,recall))
```

```python
pd.DataFrame(result, columns = ["model", "accuracy_score", "f1_score", "precision_score", "recall_score"])
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
y_pred = log_reg.predict(X_test)

acc = accuracy_score(y_test, y_pred)
fl = f1_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

result.append(("LogisticRegression",acc,fl,prec,recall))
```

```python
pd.DataFrame(result, columns = ["model", "accuracy_score", "f1_score", "precision_score", "recall_score"])
```
